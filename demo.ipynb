# Content Auto-Publish Demo
Draft content, preview the resized image, then publish.

```python
from publish_pipeline import Content, resize_and_crop, publish
from config import PLACEMENT_DIMENSIONS
from PIL import Image
```

```python
post = Content(
    title="New AI avatar workflow walkthrough",
    body="Full write-up of the workflow...",
    image_path="draft_assets/cover.png",
)
post
```

```python
resized_path = resize_and_crop(post.image_path, PLACEMENT_DIMENSIONS["blog_hero"], "draft_assets/cover_blog_hero.png")
Image.open(resized_path)
```

```python
result = publish(post, placement="blog_hero")
result["url"]
```
